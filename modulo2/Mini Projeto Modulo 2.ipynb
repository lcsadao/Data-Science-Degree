### Mini Projeto 1

## Você trabalha em uma Rede Social e precisa criar uma estrutura que represente as pessoas que a usam em python (com Programação Orientado a Objetos).
## É necessário guardar informações das pessoas que utilizam essa rede social:

nome

apelido

data de nascimento

e-mail


amigos

lista de interesses

posts


## E além disso, existem algumas funcionalidades que precisamos ter:

adicionar um amigo na lista de amigos

adicionar um interesse na lista interesses

postar

calcular a quantidade de amigos

calcular a quantidade de posts

dado um nome, verificar se existe uma pessoa com esse nome na sua lista de amizade

dado uma palavra ou frase, trazer TODOS os posts com aquela palavra (procurar post)

mostrar os posts dos amigos

## Parâmetros de correção:

Documentação
Bom nomes de classe, atributos, métodos, etc.
Lógica

```python
class User:
    
    """Creates a entity named User
        for social media purposes"""
    
    def __init__(self, name , birthday, email, nickname = None):
        
        """Constructor
            
            Parameters
        -------------------
            name:str
                User name.
            birthday:str
                User birthday.
            email:str
                User e-mail.
            nickname:str,optional
                User nickname.
        -------------------          
            """

        
        self.name = name
        self.birthday = birthday
        self.email = email
        self.nickname = nickname
        
        self.friend_list = []
        self.list_of_interests = []
        self.posts = []
        
    def add_friend(self, friend):
        
        """Method that adds 'User' objects to self.friend_list
            while adding 'self' to User.friend_list"""
        
        if friend not in self.friend_list:            

            self.friend_list.append(friend)

            friend.friend_list.append(self)

            return f'Adicionamos {friend.name} aos amigos de {self.name}!'

        else:

            return f'{friend.name} já está na sua lista de amigos!'


    
    def add_interest(self, interest):
        
        """Method to add strings named 'interest' to self.list_of_interests"""
        
        return self.list_of_interests.append(interest)        
    
    def do_post(self, post):
        
        """Method to add strings named 'post' to self.posts"""
        
        return self.posts.append(post)
    
    def friends_number(self):
        
        """Method to show the number of friends the User object has."""
        
        return f'{self.name} tem {len(self.friend_list)} amigo(s)!'
    
    def posts_number(self):
        
        """Method to show the number of posts the User object has."""
        
        return f'{self.name} postou {len(self.posts)} vez(es)!'
    
    def search_friend(self, friend):
        
        """Method to search a object from the same class in self.friend_list"""
        
        if friend in self.friend_list:
            return f'{friend.name} está na sua lista de amigos!'
        else:
            return f'{friend.name} não está na sua lista de amigos!'

    
    def search_post(self, palavra):
        
        """Method to find posts in self.posts with determined string in it."""
        
        print(f'Suas postagens com a(s) palavra(s) "{palavra}":\n')
        
        for post in self.posts:
            if palavra in post:
                print(f'{post}\n')

    def show_friends_posts(self):
        
        """Method to show post lists from other User objects in self.friend_list"""
        
        for friend in self.friend_list:
            for post in friend.posts:
                print(f'{friend.name} postou:\n {post}\n')
            
        
#Objetos        
#----------------------------------------

Lucas = User(name = 'Lucas',
             birthday = '29/03/1996',
             email = 'lucas2614@outlook.com',
             nickname = 'Adão',
                      )



Carlos = User(name = 'Carlos',
             birthday = '22/01/1991',
             email = 'carlos11@gmail.com',
                      )


Laura = User(name = 'Laura',
             birthday = '29/07/1998',
             email = 'laura@outlook.com',
                      )

#----------------------------------------
#Testes
#----------------------------------------

#Adicionar amigo
Lucas.add_friend(Carlos)


#Procurar amigo na lista de amigos
Carlos.search_friend(Lucas)

#Procurar um amigo que não está na lista de amigos
Carlos.search_friend(Laura)

#Adicionar á lista de interesses
Lucas.add_interest('Futebol')

#Mostrando a lista após adição de interesse
Lucas.list_of_interests

#Método de postagem
Lucas.do_post('Bom dia!! Boa semana para todos!')
Lucas.do_post('Ótimo dia de trabalho! Muito gratificante!')

#Método de postagem (2)
Laura.do_post('Ótima segunda feira para um passeio! Bom dia!')

#Calcular número de amigos
Lucas.friends_number()

#Calcular número de postagens
Lucas.posts_number()

#Procura de postagens com determinada string
Lucas.search_post('dia')

#Postagem para teste do método show_friends_posts
Carlos.do_post('Boa noite!')

#Adição de amigo para teste de método show_friends_posts
Lucas.add_friend(Laura)

#Postagens de amigos do objeto
Lucas.show_friends_posts()
```
