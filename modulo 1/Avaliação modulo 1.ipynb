# Avaliação módulo 1 - Degree Data Science

Na empresa em que você trabalha foi solicitada a implementação de um sistema de cadastro dos fornecedores. As informações que devem ser cadastradas são:

- Código do fornecedor;
- Nome do fornecedor;
- Telefone do fornecedor;
- Email do fornecedor.

Para isso, você deverá criar um script para armazenar essas informações em alguma estrutura de dados.

Também como parte do sistema de cadastro, é necessário criar uma "interface de navegação" pelo sistema, que tenha as seguintes funcionalidades:

- Permite a exibição (usando o `print`) de todas as informaçoes de contatos específicos, a partir da busca pelo código do fornecedor;
- Permite a remoção e adição de fornecedores à estrutura de cadastro.

Requisitos:

- Requisito #1 - Não existe um limite para fornecedores cadastrados, portanto você deverá permitir que o usuário cadastre a quantidade que desejar.
- Requisito #2 - Crie pelo menos uma função em seu script. Você deverá escolher pelo menos uma tarefa do seu script e implementar em uma função para chamá-la dentro da sua lógica de cadastro.
- Requisito #3 - Crie pelo menos uma verificação em seu código, por exemplo: Se o dicionário estiver vazio, não permita a opção remover.

Tópicos avaliados:

- Variáveis;
- Tipos de dados;
- Operadores aritméticos;
- I/O de dados;
- Expressões lógicas e condicionais;
- Laços de repetição;
- Listas e dicionários;
- Funções.


```python
def inicio():
    
    """Menu inicial com opções para direcionar o usuário ao que deseja realizar."""
    
    opcao = input('\nBem vindo á central de cadastro!\n\n Para cadastrar um fornecedor, digite 1.\n\n Para consultar um fornecedor, digite 2.\n\n Para remover um fornecedor, digite 3:\n\n')
    while (opcao != '1') and (opcao != '2') and (opcao != '3'):
        opcao = input('\nOpção inválida! Digite uma opção válida: \n\n')
    else:
        if opcao == '1':
            cadastrar()
        elif opcao == '2':
            if len(cadastro["Códigos"]) == 0:
                print('\nNão podemos consultar pois não há fornecedores registrados! Voltando ao menu inicial...\n\n')
                inicio()
            else:
                consultar(int(input('\nDigite o código do fornecedor para consultar suas informações: \n\n')))
        elif opcao == '3':
            if len(cadastro["Códigos"]) == 0:
                print('\nNão podemos remover pois não há fornecedores registrados! Voltando ao menu inicial...\n\n')
                inicio()
            else:
                remover(int(input('\nDigite o código do fornecedor que deseja remover: \n\n')))

                            
def cadastrar():
    
    """Função para cadastrar/armazenar os dados (código,nome,telefone e email) no banco de dados."""
    
    while True: 
        
        codigo = int(input('\nDigite um código para atribuir á este fornecedor (Somente números!): \n\n'))

        while codigo in cadastro["Códigos"]:
            codigo = int(input('\nCódigo já em uso! Digite outro código: \n\n'))
        else:
            cadastro["Códigos"].append(codigo)

        nome = input('\nDigite um nome para este fornecedor: \n\n')

        while nome in cadastro["Nomes"]:
            nome = input('\nNome já em uso! Digite outro nome: \n\n')
        else:
            cadastro["Nomes"].append(nome)

        telefone = int(input('\nDigite o telefone para este fornecedor (Somente números!): \n\n'))

        while telefone in cadastro["Telefones"]:
            telefone = int(input('\nTelefone já em uso! Digite outro número: \n\n'))
        else:
            cadastro["Telefones"].append(telefone)

        email = input('\nDigite um email para este fornecedor: \n\n')

        while email in cadastro["Emails"]:
            email = input('\nEmail já em uso! Digite um email: \n\n')
        else:
            cadastro["Emails"].append(email)
            break
    
        
    ask = input('\nCadastro efetuado! Deseja cadastrar mais algum fornecedor? (Digite "sim" ou "nao")\n\n')
    
    while ask != 'sim' and ask != 'nao':
        ask = input('\nDigite "sim" para cadastrar outro fornecedor ou "nao" para voltar ao menu inicial...\n\n')
    
    if ask == 'sim':
        print('\nCerto! Vamos cadastrar mais um fornecedor!\n\n')
        cadastrar()
    
    elif ask == 'nao':
        print('\nCerto! Iremos retornar ao menu inicial!\n\n')
        inicio()
    
        
def consultar(x):
    
    """Consulta o fornecedor solicitado e imprime todos os dados referentes ao código do mesmo."""
    
    while x not in cadastro["Códigos"]:
        x = int(input('\nCódigo não encontrado! Digite novamente: \n\n'))
    else:
        index_x = cadastro["Códigos"].index(x)
        print(f'\nDados do fornecedor: \n\nCódigo: {x}  \nNome: {cadastro["Nomes"][index_x]} \nTelefone: {cadastro["Telefones"][index_x]} \nEmail: {cadastro["Emails"][index_x]}\n')
        
    ask = input('\nPronto! Deseja consultar mais algum fornecedor? (Digite "sim" ou "nao"): ')
    
    while (ask != 'sim') and (ask != 'nao'):
        
        ask = input('\nDigite "sim" para consultar outro fornecedor ou "não" para voltar ao menu inicial...\n\n')
    
    else:   
        if ask == 'sim':
            if len(cadastro["Códigos"]) == 0:
                    print('\nNão podemos consultar pois não há fornecedores registrados! Voltando ao menu inicial...\n\n')
                    inicio()
            else:
                consultar(int(input('\nDigite o código do fornecedor para consultar suas informações: \n\n')))
        elif ask == 'nao':
            print('\nCerto! Vamos voltar para o menu inicial...\n\n')
            inicio()
        

def remover(x):
    
    """Remove o fornecedor solicitado pelo código referente."""
        
    while x not in cadastro["Códigos"]:
        x = int(input('\nCódigo não encontrado! Digite novamente: \n\n'))
    else:
        index_x = cadastro["Códigos"].index(x)
        cadastro["Códigos"].pop(index_x)
        cadastro["Nomes"].pop(index_x)
        cadastro["Telefones"].pop(index_x)
        cadastro["Emails"].pop(index_x)
                
    
    ask = input('\nPronto! Fornecedor removido! Deseja remover mais algum fornecedor? (Digite "sim" ou "nao"): \n\n')
    
    while (ask != 'sim') and (ask != 'nao'):
        
        ask = input('\nDigite "sim" para remover outro fornecedor e "nao" para voltar ao menu inicial: \n\n')
    
    else:    
        if (ask == 'sim'):
            if len(cadastro["Códigos"]) == 0:
                print('\n Não podemos remover pois não há fornecedores registrados! Voltando ao menu inicial...\n\n')
                inicio()
            else: 
                remover(int(input('\nDigite o código do fornecedor que deseja remover: \n\n')))
        elif (ask == 'nao'):
            print('\nCerto! Vamos retornar ao menu inicial!\n\n')
            inicio()
        


cadastro = {"Códigos" : [],
            "Nomes" : [],
            "Telefones" : [],
            "Emails" : []}

inicio()
```
